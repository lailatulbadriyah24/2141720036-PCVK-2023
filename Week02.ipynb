2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16
dan int32?

**Jawab:** Perbedaan antara int16 dan int32 adalah jumlah bit yang digunakan, dengan int 16 menggunakan 16 bit dan int32 menggunakan 32 bit. Hal ini mempengaruhi rentang nilai yang dapat direpresentasikan dalam gambar hitam. int16 mempunyai rentang lebih terbatas, sementara int32 memiliki rentang yang lebih besar. Pemilihan int16 atau int32 ini tergantung pada tingkat ketelitian yang diperlukan dan ruang penyimpanan yang tersedia.
