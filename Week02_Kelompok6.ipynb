```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

ktp = cv.imread('/content/drive/MyDrive/PCVK/ktp.jpeg')
ktp2 = cv.cvtColor(ktp,cv.COLOR_BGR2RGB) #konversi BGR -> RGB

cv.rectangle(ktp2, pt1=(190,325), pt2=(485,345), color=(244,48,100), thickness=cv.FILLED) #tutup pekerjaan
cv.rectangle(ktp2, pt1=(190,350), pt2=(300,367), color=(0,0,255), thickness=cv.FILLED) #tutup kewarganegaraan

font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(ktp2, text='CENSORED', org=(290, 340), fontFace=font, fontScale=0.5, color=(255, 255, 255), thickness=2, lineType=cv.LINE_AA) #add teks untuk pekerjaan
cv.putText(ktp2, text='CENSORED', org=(203, 364), fontFace=font, fontScale=0.5, color=(255, 255, 255), thickness=2, lineType=cv.LINE_AA) #add teks untuk kewarganegaraan

plt.imshow(ktp2)
plt.show()
```
